```python
# Drop na values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.drop(['Quantity', 'Date', 'Price', 'Country', 'Quantity', 'CustomerID'], axis = 1, inplace = True)
df.head()
```

```python
df.groupby('BillNo').count()
```

```python
df_gk = df.groupby(['BillNo', 'Itemname'])['Itemname'].count().unstack().reset_index().fillna(0).set_index('BillNo')
df_gk.head()
```

```python
def encode(x) :
    if x >= 1 :
        return 1
    if x <= 0 :
        return 0
    
    
basket = df_gk.applymap(encode)
basket
```

```python
frequent_itemsets = apriori(basket, min_support = 0.02 ,use_colnames =True)
frequent_itemsets = frequent_itemsets.sort_values(by = 'support', ascending = False)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1)
rules.sort_values(by='lift', ascending = False).head(10)
```
